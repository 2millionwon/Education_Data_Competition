```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 예제 시계열 데이터 생성
data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022,2023],
    'Value': [1.24, 1.17, 1.05, 0.98, 0.92, 0.84, 0.81, 0.78,0.72]
}
df = pd.DataFrame(data)

# 독립 변수와 종속 변수 설정
X = df[['Year']]  # 독립 변수
y = df['Value']   # 종속 변수

# 선형 회귀 모델 생성 및 학습
model = LinearRegression()
model.fit(X, y)

# 회귀 계수와 절편 출력
print(f"회귀 계수: {model.coef_[0]}")
print(f"절편: {model.intercept_}")

# 예측 값 생성
y_pred = model.predict(X)

# 2024년 추가 데이터 예측
future_years = np.array([[2024]])
future_pred = model.predict(future_years)

# 시각화
plt.figure(figsize=(10, 6))
plt.bar(df['Year'], df['Value'], color='blue', label='Actual Value')
plt.plot(df['Year'], y_pred, color='red', marker='o', linestyle='-', label='Predicted Value')

# 2024년 예측 값 추가
plt.plot(future_years, future_pred, color='green', marker='o', linestyle='-', label='2024 Predicted Value')
plt.scatter(future_years, future_pred, color='green', s=100)

# 각 막대 위에 값 표시
for i, v in enumerate(df['Value']):
    plt.text(df['Year'][i], v + 0.05, f'{v:.2f}',fontsize=8, ha='center')

# 2024년 예측 값 텍스트로 표시
for i, txt in enumerate(future_pred):
    plt.text(future_years[i], future_pred[i]+ 0.03, f'{txt:.2f}', fontsize=12, color='green', ha='center')

plt.xticks(df['Year'])
plt.xlabel('Year')
plt.ylabel('Count')
plt.title("Total Fertility Rate (TFR)")
plt.legend()
plt.show()

# 2024년 예측 값 출력
print(f"2024년 예측 값: {future_pred[0]}")
```
