```python
df_b= pd.read_excel('C:/Users/chldy/OneDrive/바탕 화면/용용/딥러닝 부트캠프/교육데이터공모전/2023년 하반기 유초중등 학교별 학년별 학생수 학급수 교원 직원_240221.xlsx',sheet_name='학교별 주요 통계', header=12, usecols='A:CV')
df_a= pd.read_excel('C:/Users/chldy/Downloads/2023년 유초중등 학교별 학년별 학생수 학급수 입학 졸업 교원 직원 면적_240319 (1).xlsx',sheet_name='학교별 주요통계', header=14, usecols='A:ER')
df_comb= pd.read_excel('C:/Users/chldy/Downloads/final_df.xlsx')
```

```python
comb_list= df_comb['학교명'].to_list()
```

```python
df_ex_comb= df_b[~df_b['학교명'].isin(comb_list)]
```

```python
df_ex_comb # 통합학교 뺌
```

```python
df_elementary= df_ex_comb[(df_ex_comb['시도']=='인천')&(df_ex_comb['학교급']=='초등학교')] #통합학교 뺀 데이터 중 인천 초, 중, 고 (교지 면적 없음) 하반기
df_middle= df_ex_comb[(df_ex_comb['시도']=='인천')&(df_ex_comb['학교급']=='중학교')]
df_high= df_ex_comb[(df_ex_comb['시도']=='인천')&(df_ex_comb['학교급']=='고등학교')]

```

```python
df_aa_ele= df_a[(df_a['시도']=='인천')&(df_a['학교급']=='초등학교')] #통합학교 있는 교지 면적 o
df_aa_mid= df_a[(df_a['시도']=='인천')&(df_a['학교급']=='중학교')]
df_aa_hig= df_a[(df_a['시도']=='인천')&(df_a['학교급']=='고등학교')]
```

```python
df_elementary
```

```python
df_aa= df_aa_ele[['학교명', '전체직원_계','일반\n교실','교과\n교실','특별\n교실','수준별교실','기타\n교실','교지면적','학생\n1인당\n교지면적']] #교지 면적 o, 상반기
df_bb= df_aa_mid[['학교명', '전체직원_계','일반\n교실','교과\n교실','특별\n교실','수준별교실','기타\n교실','교지면적','학생\n1인당\n교지면적']]
df_cc= df_aa_hig[['학교명', '전체직원_계','일반\n교실','교과\n교실','특별\n교실','수준별교실','기타\n교실','교지면적','학생\n1인당\n교지면적']]
```

```python
df_aa
```

```python
merge_df_elementary= pd.merge(df_elementary, df_aa,how='inner',on='학교명')
merge_df_middle= pd.merge(df_middle, df_bb,how='inner',on='학교명')
merge_df_high= pd.merge(df_high, df_cc,how='inner',on='학교명')
```

```python
merge_df_elementary
```

```python
merge_df_elementary.to_csv('incheon_elementary.csv')
merge_df_middle.to_csv('incheon_middle.csv')
merge_df_high.to_csv('incheon_high.csv')
```
